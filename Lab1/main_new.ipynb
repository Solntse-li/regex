```python
import Avtomat_sm
 
class AppClass:
 
    def __init__(self):
        self._fsm = Avtomat_sm.AppClass_sm(self)
        self.is_valid = False
        self.counter = 0
        self.server_end = 0
        self.server = ""
 
    def ResetServer(self):
        self.server = ""
        self.server_end = 0
 
    def Server(self):
        return self.server
 
    def IncCounter(self):
        self.counter += 1
 
    def ResetCounter(self):
        self.counter = 0
 
    def Valid(self):
        self.is_valid = True
 
    def Invalid(self):
        self.is_valid = False
 
    def check(self, string):
        ResetCounter()
        ResetServer()
        Invalid()
        self._fsm.enterStartState()
        if(len(string) > 63):
            return False
        for char in string[:6]:
            self._fsm.nfs(self.counter, char)
        for char in string[6:]:
            self._fsm.CheckChar(self.counter, char)
            if char == '/':
                self.server_end = 1
            if self.server_end != 0:
                self.server += char
        self._fsm.EOS(self.counter)
        return self.is_valid
 
 
file = open('string.txt', 'r')
string = file.read()
Avtomat = AppClass()
servers = {}
while string:
    result = Avtomat.check(string)
    if result:
        if Avtomat.Server() in servers:
            servers[Avtomat.Server()] += 1
        else:
            servers[Avtomat.Server()] == 1
    print(string)
    print(result)
    string = file.read()
```
